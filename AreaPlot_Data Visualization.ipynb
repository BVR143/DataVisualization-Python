# Area Plots
An area plot, also known as stacked area plot, is a type of Data Visualization that represents a cumilative value of different variables over continuos axis

To visualize the data points as Area plot, The stackplot() function is used. The functions title(), xlabel(), and ylabel() are used to set the title, the labels of X and Y axis respectivally 

```python
import matplotlib.pyplot as plt

# sample data for Area plot also known as stacked plot 
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016]
variable1 = [2, 4, 6, 8, 10, 12, 14]
variable2 = [1, 3, 5, 7, 9, 11, 13]
variable3 = [0, 2, 4, 6, 8, 10, 12]

# ploting the stacked plot using the stackplot() function of matplotlib library
plt.stackplot(years, variable1, variable2, variable3, labels=['Variable 1', 'Variable 2', 'Variable 3'])

# Let us make the title of the plot using the function 'title()'
plt.title("Area plot Example")

# label the x and y axis of the plot using the functions 'xlabel()' and 'ylabel()' respectively 
plt.xlabel('Years')
plt.ylabel('Values')

# Display the legend on upper left that identifies each variable 
plt.legend(loc = 'upper left')

# Display the plot 
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data for monthly revenue (in thousands of dollars)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
revenue_a = [100, 120, 150, 180, 200, 220]
revenue_b = [80, 100, 130, 160, 180, 200]
revenue_c = [70, 90, 110, 130, 150, 170]

# Creating the area plot
plt.figure(figsize=(10, 6))

plt.stackplot(months, revenue_a, revenue_b, revenue_c,
              labels=['Product A', 'Product B', 'Product C'],
              alpha=0.8)

# Adding labels and title
plt.xlabel('Months')
plt.ylabel('Revenue (in thousands)')
plt.title('Monthly Revenue by Product')

# Adding legend
plt.legend(loc='upper left')

# Display the plot
plt.tight_layout()
plt.show()
```
